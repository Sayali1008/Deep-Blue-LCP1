```
import pandas as pd
import pylab as pl
import numpy as np
df = pd.read_csv('/content/drive/My Drive/Deep Blue/Data/Wardwise Data/CSV Data Files/AWard.csv', header = None)
```

```
df.head()
```

```
cols = ['Year','Connections','TotalPop','Slum','NonSlum','Demand','Consumption']
df.columns = cols
df.shape
```

```
from sklearn import preprocessing 
scale=df.iloc[:,1:7].values;
print(scale);
  
min_max_scaler = preprocessing.MinMaxScaler(feature_range =(1000, 1200))
x_after_min_max_scaler = min_max_scaler.fit_transform(scale) 
  
print ("\nAfter min max Scaling : \n", x_after_min_max_scaler) 
print(x_after_min_max_scaler)

  
Standardisation = preprocessing.StandardScaler()   
# Scaled feature
x_after_Standardisation = Standardisation.fit_transform(scale) 
print ("\nAfter Standardisation : \n", x_after_Standardisation) 
```

```
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
vif["features"] = df.columns
```

```
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing 
scale=df.iloc[:,2:7].values;

  
min_max_scaler = preprocessing.MinMaxScaler(feature_range =(4000,5000))
x_after_min_max_scaler = min_max_scaler.fit_transform(scale) 
print(x_after_min_max_scaler)
i=0
for row in x_after_min_max_scaler :
    for elem in row:
        df.at[i,'TotalPop']=elem
        df.at[i,'Slum']=elem
        df.at[i,'NonSlum']=elem
        df.at[i,'Demand']=elem
        df.at[i,'Consumption']=elem
    i+=1   
```
